Looking for the nan values

```python
# Checking for null values
df.isna().sum()
```

```python
# dropping the rows containing the null values and rewriting on same dataset
df.dropna(inplace=True)
```

```python
df.info()
```

```python
# boolean indexing is done , keepind only the rows whose instruction columns contains the 
# data not '[]'
df = df[df['instructions'].str.strip() != '[]']
```

```python
df.info()
```

```python
df=df[df['ingredients'].str.strip()!='[]']
```

```python
df.info()
```

```python
df.info()
```

```python
def clean_ingredients(ingredient_str):
    try:
        items = eval(ingredient_str)  # Convert string to list here , it treats expr inside string as executable code
        return [item.strip() for item in items if item.strip() and '\t' not in item] # remove white spaces, columns has \t in it (as there were few \t\t\t\t...)
    
    except:
        return []
```

```python
# applying the clean function to every column in the ingredients column
df['ingredients'] = df['ingredients'].apply(clean_ingredients)
```
